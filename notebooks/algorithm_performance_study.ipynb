```python
import numpy as np

sizes = [50, 100, 200, 400]
avg_comparisons = []

trials = 10

for size in sizes:
    trial_results = []
    for _ in range(trials):
        data = [random.randint(1, 1000) for _ in range(size)]
        result = bubble_sort(data)
        trial_results.append(result.comparisons)
    avg_comparisons.append(np.mean(trial_results))

plt.plot(sizes, avg_comparisons)
plt.xlabel("Input Size")
plt.ylabel("Average Comparisons")
plt.title("Bubble Sort Average Complexity")
plt.show()
```

```python
import random
import matplotlib.pyplot as plt
```

```python
sizes = [50, 100, 200, 400]

random_comparisons = []
sorted_comparisons = []

for size in sizes:
    # Random input
    data_random = [random.randint(1, 1000) for _ in range(size)]
    result_random = bubble_sort(data_random)
    random_comparisons.append(result_random.comparisons)
    
    # Sorted input
    data_sorted = list(range(size))
    result_sorted = bubble_sort(data_sorted)
    sorted_comparisons.append(result_sorted.comparisons)

plt.plot(sizes, random_comparisons, label="Random Input")
plt.plot(sizes, sorted_comparisons, label="Sorted Input")

plt.xlabel("Input Size")
plt.ylabel("Number of Comparisons")
plt.title("Bubble Sort: Random vs Sorted")
plt.legend()
plt.show()
```

```python
sizes = [50, 100, 200, 400]

bubble_results = []
insertion_results = []

for size in sizes:
    data = [random.randint(1, 1000) for _ in range(size)]
    
    bubble = bubble_sort(data)
    insertion = insertion_sort(data)
    
    bubble_results.append(bubble.comparisons)
    insertion_results.append(insertion.comparisons)

plt.plot(sizes, bubble_results, label="Bubble Sort")
plt.plot(sizes, insertion_results, label="Insertion Sort")

plt.xlabel("Input Size")
plt.ylabel("Number of Comparisons")
plt.title("Bubble vs Insertion (Random Input)")
plt.legend()
plt.show()
```
