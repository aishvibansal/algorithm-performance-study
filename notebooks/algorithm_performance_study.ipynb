```python
import time
import numpy as np

def measure_time(func, data):
    start = time.perf_counter()
    func(data)
    end = time.perf_counter()
    return end - start
```

```python
import time
import numpy as np

def measure_time(func, data):
    start = time.perf_counter()
    func(data)
    end = time.perf_counter()
    return end - start
```

```python
sizes = [10, 20, 50, 100, 200, 400]
trials = 20

insertion_times = []
merge_times = []

for size in sizes:
    insertion_trial = []
    merge_trial = []
    
    # Generate ONE dataset per size
    base_data = [random.randint(1, 1000) for _ in range(size)]
    
    for _ in range(trials):
        insertion_trial.append(measure_time(insertion_sort, base_data))
        merge_trial.append(measure_time(merge_sort, base_data))
    
    insertion_times.append(np.mean(insertion_trial))
    merge_times.append(np.mean(merge_trial))

```

```python
plt.plot(sizes, insertion_times, label="Insertion Sort")
plt.plot(sizes, merge_times, label="Merge Sort")

plt.xlabel("Input Size")
plt.ylabel("Execution Time (seconds)")
plt.title("Runtime Comparison: Insertion vs Merge")
plt.legend()
plt.show()
```

```python
sizes = [50, 100, 200, 300, 400, 500]

first_pivot_comparisons = []
random_pivot_comparisons = []

for size in sizes:
    data = list(range(size))  # Sorted input (worst-case for first pivot)

    first_result = quicksort_first_pivot(data)
    random_result = quicksort_random_pivot(data)

    first_pivot_comparisons.append(first_result.comparisons)
    random_pivot_comparisons.append(random_result.comparisons)
```

```python
plt.plot(sizes, first_pivot_comparisons, marker='o', label="Quicksort (First Pivot)")
plt.plot(sizes, random_pivot_comparisons, marker='o', label="Quicksort (Random Pivot)")

plt.xlabel("Input Size")
plt.ylabel("Number of Comparisons")
plt.title("Pivot Sensitivity: Sorted Input")
plt.legend()
plt.show()
```

```python
for i in range(len(sizes)):
    print(f"Size {sizes[i]} | First Pivot: {first_pivot_comparisons[i]} | Random Pivot: {random_pivot_comparisons[i]}")
```
